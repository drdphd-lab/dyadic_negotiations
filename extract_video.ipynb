### EDA

```python
# Время видео, в котором по факту игрались раунды
general_info = pd.read_pickle('data/general_info.pkl')
round_dur_str = list(
    zip(
        general_info['round_1_time'].map(lambda x: x[0]),
        general_info['round_1_time'].map(lambda x: x[1]),
        general_info['round_2_time'].map(lambda x: x[0]),
        general_info['round_2_time'].map(lambda x: x[1])
    )
)

round_1_dur_dct = {}
round_2_dur_dct = {}
for idx, row in general_info.iterrows():
    round_1_start = int(row['round_1_time'][0][:2])*60 + int(row['round_1_time'][0][3:])
    round_1_end = int(row['round_1_time'][1][:2])*60 + int(row['round_1_time'][1][3:])
    
    round_2_start = int(row['round_2_time'][0][:2])*60 + int(row['round_2_time'][0][3:])
    round_2_end = int(row['round_2_time'][1][:2])*60 + int(row['round_2_time'][1][3:])
    
    round_1_dur_dct[row['title']] = (round_1_start, round_1_end)
    round_2_dur_dct[row['title']] = (round_2_start, round_2_end)
```

```python
emotion_series = pd.Series()
for title in tqdm(general_info.loc[general_info['issue'] == '-', 'title']):
    
    left, right = split_players(emotion_dct[title])
    emotions = {idx: em for idx, em in enumerate(pd.DataFrame(left).columns.tolist())}
    
    one_emotion_left = (
        pd.DataFrame(left)
        .drop(
            pd.RangeIndex(
                start=round_1_dur_dct[title][1]*25, stop=round_2_dur_dct[title][0]*25
            ), axis=0
        )
        .dropna()
        .apply(np.argmax, axis=1)
        .map(emotions)
    )
    
    one_emotion_right = (
        pd.DataFrame(right)
        .drop(
            pd.RangeIndex(
                start=round_1_dur_dct[title][1]*25, stop=round_2_dur_dct[title][0]*25
            ), axis=0
        )
        .dropna()
        .apply(np.argmax, axis=1)
        .map(emotions)
    )
    
    emotion_series = pd.concat(
        [one_emotion_left, one_emotion_right, emotion_series],
        axis=0, ignore_index=True
    )
```

Эмоциональная нестабильность

```python
from collections import defaultdict

unstable_count_overall = []

for title in tqdm(general_info.loc[general_info['issue'] == '-', 'title']):
    
    left, right = split_players(emotion_dct[title])
    emotions = {idx: em for idx, em in enumerate(pd.DataFrame(left).columns.tolist())}
    
    one_emotion_left = (
        pd.DataFrame(left)
        .drop(
            pd.RangeIndex(
                start=round_1_dur_dct[title][1]*25, stop=round_2_dur_dct[title][0]*25
            ), axis=0
        )
        .apply(np.argmax, axis=1)
        .map(emotions)
    )
    
    prev = 0
    unstable_count_left = 0
    em_len_counter_left = 0
    emotion_len_left = defaultdict(lambda: [])
    for em in one_emotion_left:
        em_len_counter_left += 1
        if em != prev:
            emotion_len_left[em].append(em_len_counter_left)
            em_len_counter_left=0
            unstable_count_left += 1
            prev = em
    unstable_count_left_share = unstable_count_left / one_emotion_left.shape[0]
    unstable_count_overall.append(unstable_count_left_share)
    
    
    one_emotion_right = (
        pd.DataFrame(right)
        .drop(
            pd.RangeIndex(
                start=round_1_dur_dct[title][1]*25, stop=round_2_dur_dct[title][0]*25
            ), axis=0
        )
        .apply(np.argmax, axis=1)
        .map(emotions)
    )
    
    prev = 0
    unstable_count_right = 0
    em_len_counter_right = 0
    emotion_len_right = defaultdict(lambda: [])
    for em in one_emotion_right:
        em_len_counter_right += 1
        if em != prev:
            emotion_len_right[em].append(em_len_counter_right)
            em_len_counter_right=0
            unstable_count_right += 1
            prev = em
    
    unstable_count_right_share = unstable_count_right / one_emotion_right.shape[0]
    unstable_count_overall.append(unstable_count_right_share)
```

```python
em_df = pd.DataFrame()
for key, value in emotion_len_left.items():
    em_df = pd.concat([em_df, pd.DataFrame({'em': key, 'count': value})], axis=0)
```

### Гипотеза - нейтральные/положительые эмоции

```python
from scipy.stats import ttest_ind
```

```python
neutral_positive_dct = {
    'judge_1': {'winners': [], 'losers': []},
    'judge_2': {'winners': [], 'losers': []},
    'judge_3': {'winners': [], 'losers': []},
    'judge_4': {'winners': [], 'losers': []},
    'titles': []
}
```

```python
for title in tqdm(general_info.loc[general_info['issue'] == '-', 'title']):
    neutral_positive_dct['titles'].append(title)
    left, right = split_players(emotion_dct[title])
    emotions = {idx: em for idx, em in enumerate(pd.DataFrame(left).columns.tolist())}
    
    one_emotion_left = (
        pd.DataFrame(left)
        .dropna()
        .apply(np.argmax, axis=1)
        .map(emotions)
        .value_counts()
    )
    share_left = one_emotion_left / one_emotion_left.sum()
    left_np = share_left['happy'] + share_left['neutral']
    
    one_emotion_right = (
        pd.DataFrame(right)
        .dropna()
        .apply(np.argmax, axis=1)
        .map(emotions)
        .value_counts()
    )
    share_right = one_emotion_right / one_emotion_right.sum()
    right_np = share_right['happy'] + share_left['neutral']
    
    left_pos = videos_mapping_dct[title][0]
    right_pos = videos_mapping_dct[title][1]
    for i in range(4):
        cond = (
            general_info.loc[
                general_info['title'] == title, 'judge_decision'
            ].values[0][i] == 'player_1'
        )
        if cond:
            if left_pos == 'player_1':
                neutral_positive_dct[f'judge_{i+1}']['winners'].append(left_np)
                neutral_positive_dct[f'judge_{i+1}']['losers'].append(right_np)
            else:
                neutral_positive_dct[f'judge_{i+1}']['winners'].append(right_np)
                neutral_positive_dct[f'judge_{i+1}']['losers'].append(left_np)  
        else:
            if left_pos == 'player_2':
                neutral_positive_dct[f'judge_{i+1}']['winners'].append(left_np)
                neutral_positive_dct[f'judge_{i+1}']['losers'].append(right_np)
            else:
                neutral_positive_dct[f'judge_{i+1}']['winners'].append(right_np)
                neutral_positive_dct[f'judge_{i+1}']['losers'].append(left_np)  
```

### Гипотеза - эмоциональная нестабильность

```python
unstable_dct = {
    'judge_1': {'winners': [], 'losers': []},
    'judge_2': {'winners': [], 'losers': []},
    'judge_3': {'winners': [], 'losers': []},
    'judge_4': {'winners': [], 'losers': []},
    'titles': []
}
```

```python
for title in tqdm(general_info.loc[general_info['issue'] == '-', 'title']):
    
    unstable_dct['titles'].append(title)
    left, right = split_players(emotion_dct[title])
    emotions = {idx: em for idx, em in enumerate(pd.DataFrame(left).columns.tolist())}
    
    one_emotion_left = (
        pd.DataFrame(left)
        .drop(
            pd.RangeIndex(
                start=round_1_dur_dct[title][1]*25, stop=round_2_dur_dct[title][0]*25
            ), axis=0
        )
        .apply(np.argmax, axis=1)
        .map(emotions)
    )
    
    prev = 0
    unstable_count_left = 0
    for em in one_emotion_left:
        if em != prev:
            unstable_count_left += 1
            prev = em
    unstable_share_left = unstable_count_left / one_emotion_left.shape[0]
    
    one_emotion_right = (
        pd.DataFrame(right)
        .drop(
            pd.RangeIndex(
                start=round_1_dur_dct[title][1]*25, stop=round_2_dur_dct[title][0]*25
            ), axis=0
        )
        .apply(np.argmax, axis=1)
        .map(emotions)
    )
    
    prev = 0
    unstable_count_right = 0
    for em in one_emotion_right:
        if em != prev:
            unstable_count_right += 1
            prev = em
    
    unstable_share_right = unstable_count_right / one_emotion_right.shape[0]
    
    
    left_pos = videos_mapping_dct[title][0]
    right_pos = videos_mapping_dct[title][1]
    
    for i in range(4):
        cond = (
            general_info.loc[
                general_info['title'] == title, 'judge_decision'
            ].values[0][i] == 'player_1'
        )
        if cond:
            if left_pos == 'player_1':
                unstable_dct[f'judge_{i+1}']['winners'].append(unstable_share_left)
                unstable_dct[f'judge_{i+1}']['losers'].append(unstable_share_right)
            else:
                unstable_dct[f'judge_{i+1}']['winners'].append(unstable_share_right)
                unstable_dct[f'judge_{i+1}']['losers'].append(unstable_share_left)  
        else:
            if left_pos == 'player_2':
                unstable_dct[f'judge_{i+1}']['winners'].append(unstable_share_left)
                unstable_dct[f'judge_{i+1}']['losers'].append(unstable_share_right)
            else:
                unstable_dct[f'judge_{i+1}']['winners'].append(unstable_share_right)
                unstable_dct[f'judge_{i+1}']['losers'].append(unstable_share_left)
```
